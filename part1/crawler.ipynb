```python
import requests
```

```python
urls = 'https://www.yanyue.cn/tobacco#brands'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
```

```python
response = requests.get(urls,headers=header)
```

```python
from lxml import etree
```

```python
html = etree.HTML(response.text)
base_url = 'https://www.yanyue.cn/'
```

```python
#获取品牌对应id
base_xpath = '/html/body/div[10]/div[5]/div[3]/div/div[1]/ul/li[@class="letterList-item clearfix"]/ul/li/a/@href'
```

```python
brand_urls = []
for i in html.xpath(base_xpath):
    brand_urls.append(i.split('/')[-1])

brand_urls
```

```python
#获取产品对应id
product_ids = []
for idx, i in enumerate(brand_urls):
    brand_url = f'https://www.yanyue.cn/sort/sort/sortid/{i}/p/1'
    brand_response = requests.get(brand_url,headers=header)
    brand_html = etree.HTML(brand_response.text)
    brand_xpath = '//*[@id="left"]/nav/div[1]/span[2]/text()'
    
    # print(brand_html.xpath(brand_xpath))
    pages = brand_html.xpath(brand_xpath)
    # print(pages)
    if len(pages) == 0:
        product_xpath = '//*[@id="prowrap"]/div/div/a/@href'
        product_id = brand_html.xpath(product_xpath)
        # print(comment)
        product_ids.extend(product_id)
    else:
        for j in range(int(pages[0])):
            product_urls = f'https://www.yanyue.cn/sort/sort/sortid/{i}/p/{j+1}'
            product_xpath = '//*[@id="prowrap"]/div/div/a/@href'
            product_response = requests.get(product_urls,headers=header)
            product_html = etree.HTML(product_response.text)
            product_id = product_html.xpath(product_xpath)
            # print(classify_html.xpath(classify_xpath))
            product_ids.extend(product_id)
    # break
    print(f'{idx}/{len(brand_urls)} {len(product_ids)}is complete')
```

```python
len(set(product_ids))
with open('product_ids.txt', 'w', encoding='utf-8') as file:
    file.write('\n'.join(list(set(product_ids))) )
    # 2953
```
