```python
def extract_str(s):
    anchor_str = "分类标签"
    start_index = s.find(anchor_str) + len(anchor_str) + 1
    first_index = s.find("\"", start_index)
    end_index = s.find("\"", first_index+1)
    return s[first_index+1:end_index]
```

```python
final_data['answer1'] = final_data['assistant1'].map(extract_str)
```

```python
final_data
```

```python
def extract_str2(s):
    anchor_str = "\"分类标签\":"
    start_index = s.find(anchor_str) 
    if start_index != -1:
        start_index += len(anchor_str)
        first_index = s.find("\"", start_index)
        end_index = s.find("\"", first_index+1)
        return s[first_index+1:end_index]
    else:
        a = 'A正面'
        b = 'B负面'
        c = 'C中性'
        d = 'D讽刺'
        e = 'E无产品评价语意'
        if a in s:
            return a
        elif b in s:
            return b
        elif c in s:
            return c
        elif d in s:
            return d
        elif e in s:
            return e
        else:
            return '未找到对应选项'
```

```python
final_data['answer2'] = final_data['assistant2'].map(extract_str2)
```
